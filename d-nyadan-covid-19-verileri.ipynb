```python
import pandas as pd # Veri çerçeveleri için
import numpy as np  # Sayısal işlemler için
import matplotlib.pyplot as plt # Temel görselleştirmeler için
import seaborn as sns # Gelişmiş görselleştirmeler için

# Grafiklerin doğrudan notebook içinde görünmesini sağlar
%matplotlib inline
```

```python
# Veri setinizin dosya adını kontrol edin (örn: full_grouped.csv, latest.csv vb.)
# Kaggle'daki veri seti sayfasında 'Data' sekmesinden doğru dosya adını bulabilirsiniz.
file_path = '/kaggle/input/covid19-data-from-world/owid-covid-data.csv' # Örnek bir dosya adı
# veya
# file_path = '../input/covid19-data-from-world/latest.csv' # Diğer bir olası dosya adı

df = pd.read_csv(file_path)

# Veri setinin ilk 5 satırını görüntüleyin
print("Veri Setinin İlk 5 Satırı:")
print(df.head())
```

```python
# Veri setinin genel bilgilerini (sütunlar, veri tipleri, eksik değerler) gösterin
print("\nVeri Seti Bilgileri:")
df.info()

# Sayısal sütunların istatistiksel özetini görüntüleyin
print("\nSayısal Sütunların İstatistiksel Özeti:")
print(df.describe())

# Her sütundaki eksik değer sayısını kontrol edin
print("\nHer Sütundaki Eksik Değer Sayısı:")
print(df.isnull().sum())

# Sütun isimlerini görüntüleyin
print("\nSütun İsimleri:")
print(df.columns)
```

```python
# 'date' sütununu datetime formatına dönüştürün
if 'date' in df.columns: # 'Date' yerine 'date'
    df['date'] = pd.to_datetime(df['date'])
    print("\n'date' sütunu datetime formatına dönüştürüldü.")
else:
    print("\n'date' sütunu bulunamadı. Lütfen sütun adını kontrol edin.")

# Eksik değerleri doldurma veya silme stratejileri
# Çıktınızda continent sütununda çok fazla eksik değer olduğunu gördüm (12673).
# total_cases, total_deaths gibi sütunlarda da eksik değerler var.
# Bu eksik değerleri nasıl ele alacağınız analizinize bağlıdır.
# Örneğin, 0 ile doldurmak yerine o satırları düşürmeyi tercih edebilirsiniz,
# ya da bazı sütunları analizden çıkarabilirsiniz.

# Genel bir örnek:
# Sayısal sütunlardaki NaN değerleri 0 ile doldurmak (dikkatli olun, bu her zaman uygun olmayabilir)
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(0)

# Veya önemli analiz sütunlarındaki (örn: total_cases) eksik değerleri içeren satırları düşürmek
# df.dropna(subset=['total_cases', 'total_deaths'], inplace=True)

# Eksik değerler sonrası kontrol
print("\nEksik Değerler (Ön İşleme Sonrası):")
print(df.isnull().sum())
```

```python
# 'date' sütununu datetime formatına dönüştürün
if 'date' in df.columns: # 'Date' yerine 'date'
    df['date'] = pd.to_datetime(df['date'])
    print("\n'date' sütunu datetime formatına dönüştürüldü.")
else:
    print("\n'date' sütunu bulunamadı. Lütfen sütun adını kontrol edin.")

# Eksik değerleri doldurma veya silme stratejileri
# Çıktınızda continent sütununda çok fazla eksik değer olduğunu gördüm (12673).
# total_cases, total_deaths gibi sütunlarda da eksik değerler var.
# Bu eksik değerleri nasıl ele alacağınız analizinize bağlıdır.
# Örneğin, 0 ile doldurmak yerine o satırları düşürmeyi tercih edebilirsiniz,
# ya da bazı sütunları analizden çıkarabilirsiniz.

# Genel bir örnek:
# Sayısal sütunlardaki NaN değerleri 0 ile doldurmak (dikkatli olun, bu her zaman uygun olmayabilir)
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(0)

# Veya önemli analiz sütunlarındaki (örn: total_cases) eksik değerleri içeren satırları düşürmek
# df.dropna(subset=['total_cases', 'total_deaths'], inplace=True)

# Eksik değerler sonrası kontrol
print("\nEksik Değerler (Ön İşleme Sonrası):")
print(df.isnull().sum())
```

```python
# Günlük toplam vakaları ve ölümleri hesaplayın
if 'date' in df.columns: # 'Date' yerine 'date'
    # Tarihe göre gruplayıp sayısal sütunları topluyoruz
    daily_totals = df.groupby('date')[['total_cases', 'total_deaths']].sum().reset_index()

    plt.figure(figsize=(15, 7))
    plt.plot(daily_totals['date'], daily_totals['total_cases'], label='Toplam Vakalar', color='blue')
    plt.plot(daily_totals['date'], daily_totals['total_deaths'], label='Toplam Ölümler', color='red')
    plt.title('Küresel COVID-19 Vaka ve Ölüm Eğilimleri')
    plt.xlabel('Tarih')
    plt.ylabel('Sayı')
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print("\n'date' sütunu bulunamadığı için zaman serisi analizi yapılamadı.")
```

```python
# Örnek: Son tarihi al
df['date'] = pd.to_datetime(df['date'])
latest_date = df['date'].max()

# O tarihteki verileri al (filtreleme sebebiyle boş kalabilir)
latest_df = df[df['date'] == latest_date]

print(latest_df[['location', 'total_cases', 'total_deaths']].head())
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(by=['location', 'date'])
latest_df = df.groupby('location').last().reset_index()

latest_df['Mortality_Rate'] = latest_df['total_deaths'] / latest_df['total_cases']
latest_df = latest_df.dropna(subset=['Mortality_Rate'])
latest_df = latest_df[latest_df['Mortality_Rate'] > 0]

top_10_mortality = latest_df.sort_values(by='Mortality_Rate', ascending=False).head(10)
print(top_10_mortality[['location', 'total_cases', 'total_deaths', 'Mortality_Rate']])
```

```python
# Günlük yeni vaka ve ölüm eğilimleri
plt.figure(figsize=(15, 7))
# Tarihe göre gruplayıp yeni vaka ve ölümleri toplayın
daily_new_totals = df.groupby('date')[['new_cases', 'new_deaths']].sum().reset_index()

plt.plot(daily_new_totals['date'], daily_new_totals['new_cases'], label='Günlük Yeni Vakalar', color='green')
plt.plot(daily_new_totals['date'], daily_new_totals['new_deaths'], label='Günlük Yeni Ölümler', color='purple')
plt.title('Küresel Günlük Yeni COVID-19 Vaka ve Ölüm Eğilimleri')
plt.xlabel('Tarih')
plt.ylabel('Sayı')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Kıta bazında son durum
if 'continent' in df.columns:
    latest_date = df['date'].max()
    df_latest_continent = df[df['date'] == latest_date].copy()
    
    # Her kıta için en son toplam değerleri al (NaN değerleri 0 ile doldurulmuştur)
    continent_summary = df_latest_continent.groupby('continent')[['total_cases', 'total_deaths']].sum().reset_index()
    
    print("\nKıta Bazında Son Durum:")
    print(continent_summary.sort_values(by='total_cases', ascending=False))

    plt.figure(figsize=(14, 7))
    sns.barplot(x='total_cases', y='continent', data=continent_summary.sort_values(by='total_cases', ascending=False), palette='plasma')
    plt.title('Kıtalara Göre Toplam COVID-19 Vakaları')
    plt.xlabel('Toplam Vaka Sayısı')
    plt.ylabel('Kıta')
    plt.show()

    plt.figure(figsize=(14, 7))
    sns.barplot(x='total_deaths', y='continent', data=continent_summary.sort_values(by='total_deaths', ascending=False), palette='cividis')
    plt.title('Kıtalara Göre Toplam COVID-19 Ölümleri')
    plt.xlabel('Toplam Ölüm Sayısı')
    plt.ylabel('Kıta')
    plt.show()
else:
    print("\n'continent' sütunu bulunamadı. Kıta bazında analiz yapılamadı.")
```

```python
# Pozitiflik Oranı ve Test Sayısı Korelasyonu (Örnek Ülkeler için)
# Sadece yeterli veriye sahip ülkeleri filtreleyebiliriz
countries_to_plot = ['United States', 'India', 'Brazil', 'United Kingdom', 'Turkey']

plt.figure(figsize=(15, 8))
for country in countries_to_plot:
    country_df = df[df['location'] == country].dropna(subset=['positive_rate', 'new_tests_smoothed_per_thousand'])
    if not country_df.empty:
        plt.plot(country_df['date'], country_df['positive_rate'], label=f'{country} Pozitiflik Oranı')
plt.title('Seçili Ülkelerde Pozitiflik Oranı Eğilimleri')
plt.xlabel('Tarih')
plt.ylabel('Pozitiflik Oranı')
plt.legend()
plt.grid(True)
plt.show()

# Not: Test başına düşen vaka oranı için de benzer bir analiz yapabilirsiniz
```
